```
import numpy as np

```

------------------------------

* SciPy: is a Python library built on top of NumPy that provides additional functionality for scientific computing and data analysis. It is an open-source library, and is built to work with NumPy arrays. SciPy provides many of the numerical routines that NumPy does not provide, such as optimization, signal processing, statistical functions, linear algebra, and more.

    * The `scipy.stats` module in Python provides a wide range of probability distributions and statistical functions. One common use case is to use it to conduct statistical tests on data sets.

```
from scipy import stats as st

```

## Example Of scipy

````python
from scipy import stats as st

data = [0.5, 0.7, 1.0, 1.2, 1.3, 2.1]

stat, p = st.shapiro(data)

print("Statistics=%.3f, p=%.3f" % (stat, p))



alpha = 0.05

if p > alpha:
    print("Sample looks Gaussian (fail to reject H0)")
else:
    print("Sample does not look Gaussian (reject H0)")
````

````python
from scipy import stats as st

data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [1.142, -0.432, -0.938, -0.729, -0.846, -0.157, 0.500, 1.183, -1.075, -0.169]

t, p = st.ttest_ind(data1, data2)

print("t = ", t)
print("p = ", p)



alpha = 0.05

if p > alpha:
    print("Same distribution (fail to reject H0)")
else:
    print("Different distribution (reject H0)")
````

## 2. Numpy

### Numpy array

````python
## remember the length of the array should be equal in size

ArrVar = np.array([[0, 1, 2, 3, 4],
                   [5, 8, 7, 8, 9],
                   [10, 8, 12, 13, 14]])
````

````python
ArrVar
````

````python
## printing first raw of the numpy array in x axis

print(ArrVar[0,:])

## printing second raw of the numpy array in x axis

print(ArrVar[1,:])

## printing third raw of the numpy array in x axis

print(ArrVar[2,:])

## printing first column of the numpy array in y axis

print(ArrVar[:,0])

## printing second column of the numpy array in y axis

print(ArrVar[:,1])

## printing third column of the numpy array in y axis

print(ArrVar[:,2])

## printing four column of the numpy array in y axis

print(ArrVar[:,3])

## printing five column of the numpy array in y axis

print(ArrVar[:,4])
````

### numpy reshape() Function

````python
## printing 15 value in the numpy array (ArrVar) using reshape() function

np.reshape(ArrVar, 15)
````

````python
## printing 3 raw and 5 columns of the numpy array (ArrVar)

np.reshape(ArrVar, (3, 5))
````

### numpy min() & max() Function

````python
print("-------------------- x-axis --------------------")

## printing maximum number of value in all 3 raw

print(np.max(ArrVar, axis = 1)) ## when axis=1 it would take the maximum of each row, but when axis=0 it would take the maximum row

## printing minimum number of value in all 3 raw

print(np.min(ArrVar, axis = 1)) 


print("-------------------- y-axis --------------------")

## printing maximum number of value in all 5 columns

print(np.max(ArrVar, axis = 0)) 

## printing minimum number of value in all 5 columns

print(np.min(ArrVar, axis = 0)) 
````

### numpy mean() & median() Function

````python
print("-------------------- x-axis --------------------")

## printing the mean of all 3 raw

print(np.mean(ArrVar, axis = 1))


## printing the median of all 3 raw

print(np.median(ArrVar, axis = 1))



print("-------------------- y-axis --------------------")

## printing the mean of all 5 columns

print(np.mean(ArrVar, axis = 0))


## printing the median of all 5 columns

print(np.median(ArrVar, axis = 0))
````

## 2. Numpy with Scipy

### numpy with scipy to find mode and standard division ( mode() & std() ) Function

````python
print("-------------------- x-axis --------------------")

mode_n = st.mode(ArrVar, axis = 1) ## using scipy to find mode
print(mode_n)

varr = np.std(ArrVar, axis = 1) ## finding standerd deviation
print(varr)


print("-------------------- y-axis --------------------")

mode_n = st.mode(ArrVar, axis = 0) ## using scipy to find mode
print(mode_n)

varr = np.std(ArrVar, axis = 0) ## finding standerd deviation
print(varr)
````

### numpy with scipy for finding variance ( var() )

````python
print("-------------------- x-axis --------------------")

print(np.var(ArrVar, axis = 1)) ## varinace


print("-------------------- y-axis --------------------")

print(np.var(ArrVar, axis = 0)) ## varinace
````

### random number generate using numpy

````python
'''

seed is a value used to initialize a pseudorandom number generator. 
The seed value is used to initialize the state of the generator, 
which is then used to generate a sequence of random numbers. 
The same seed value will always produce the same sequence of random numbers, 
so it can be used to generate reproducible results.

'''



rng = np.random.default_rng(seed = 42) ## for creating random variables 

## creating 100 raw and 50 columns
x = rng.random((100, 50))

print(x)
print("length of the x: ", len(x))
````

### saving the data in file that name is "outfile.npy"

````python
np.save('outfile', x)
b = np.load('outfile.npy') 
print(b)
````
